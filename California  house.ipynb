```python
from __future__ import print_function

import math


from matplotlib import cm
from matplotlib import gridspec
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics 
```

```python
house = pd.read_csv("housing.csv")
house.head()
```

```python
house.describe()
```

```python
house.corr()
```

```python
house.isnull().sum()
```

```python
median = house['total_bedrooms'].median()
house['total_bedrooms'].fillna(median, inplace=True)
```

```python
house.isnull().sum()
```

```python
df_object = house.select_dtypes(include=['object']).copy()
df_object.head()
```

```python
df_float = house.select_dtypes(include=['float64']).copy()
df_float
```

```python
from sklearn.preprocessing import LabelBinarizer
encode = LabelBinarizer()
df_cat = encode.fit_transform(df_object)
```

```python
cat_value = pd.DataFrame(df_cat)
```

```python
df_house = pd.concat([df_float, cat_value], axis=1)
```

```python
df_house.head()
```

```python
label = df_house['median_house_value']
df_new = df_house.drop(['median_house_value'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df_new
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(df_new ,label, test_size=0.2, random_state=42)
clf = LinearRegression()
clf.fit(Xtrain,ytrain)
accuracy = clf.score(Xtest,ytest)
accuracy
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = clf.predict(Xtest)
lin_mse = mean_squared_error(ytest, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```
